#1 - Potencial de Ativação - Neurônio Único

```python
# Potencial de Ativação - Neurônio Único (n=3)
import math
import matplotlib.pyplot as plt

# Calcula o potencial de ativação: u = sum(Wi * Xi) + bias
def activation_potential(X, W, bias):
    u = sum(w * x for w, x in zip(W, X)) + bias
    return u

X1 = [1, 2, 3]
W1 = [0.2, 0.8, -0.5]
bias1 = 2
u1 = activation_potential(X1, W1, bias1)
print(f"X     = {X1}")
print(f"W     = {W1}")
print(f"bias  = {bias1}")
print(f"Saída = {round(u1, 4)}")  # Esperado: 2.3
```

# 2 - Potencial de Ativação - Neurônio Único (n=4)

```python
# Potencial de Ativação - Neurônio Único (n=4)
import math
import matplotlib.pyplot as plt

def activation_potential(X, W, bias):
    u = sum(w * x for w, x in zip(W, X)) + bias
    return u

X1 = [1, 2, 3, 2.5]
W1 = [0.2, 0.8, -0.5, 1]
bias1 = 2
u1 = activation_potential(X1, W1, bias1)
print(f"X     = {X1}")
print(f"W     = {W1}")
print(f"bias  = {bias1}")
print(f"Saída = {round(u1, 4)}")  # Esperado: 4.8
```

# 3 - Potencial de Ativação - Camada de Neurônios

```python
# Potencial de Ativação - Camada de Neurônios
import math
import matplotlib.pyplot as plt

# Retorna uma lista dos potenciais de ativação para cada neurônio na camada.
def activation_potential(X, W_layer, bias_layer):
    activation_potentials = []
    for W_neuron, bias_neuron in zip(W_layer, bias_layer):
        u = sum(w * x for w, x in zip(W_neuron, X)) + bias_neuron
        activation_potentials.append(round(u, 4))
    return activation_potentials

X1 = [1, 2, 3, 2.5]
W1 = [0.2, 0.8, -0.5, 1]
W2 = [0.5, -0.91, 0.26, -0.5]
W3 = [-0.26, -0.27, 0.17, 0.87]
bias1 = 2
bias2 = 3
bias3 = 0.5

# Prepara inputs
W_layer = [W1, W2, W3]
bias_layer = [bias1, bias2, bias3]

# Calcula potenciais para a camada
u_layer = activation_potential(X1, W_layer, bias_layer)

# Exibe os resultados
print(f"Potenciais de Ativação da Camada: {u_layer}") # Esperado: [4.8, 1.21, 2.385]
```

# 4 - Função Degrau Bipolar Aproximada

```python
# Função Degrau Bipolar Aproximada
import numpy as np
import matplotlib.pyplot as plt

title = "Função Degrau Bipolar Aproximada"

# Gerar valores de x de -5.0 a 5.0 com passo de 0.1
x_values_bipolar = np.linspace(-5.0, 5.0, 101)

# Calcular os valores de y usando a função bipolar
y_values_bipolar = [np.sign(x) for x in x_values_bipolar]

# Plotar o gráfico
plt.figure(figsize=(10, 6))
plt.plot(x_values_bipolar, y_values_bipolar, label=title)
plt.title(f'Gráfico da {title}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.ylim(-1.2, 1.2) # A função bipolar varia entre -1 e 1
plt.yticks([-1, 0, 1])
plt.legend()
plt.show()
```

# 5 - Função Rampa Simétrica

```python
# Função Rampa Simétrica
import numpy as np
import matplotlib.pyplot as plt

title = "Função Rampa Simétrica"

# Valores de x de -5.0 a 5.0 com passo de 0.1
x_values = np.linspace(-5.0, 5.0, 101)

# Valores de y usando a função degrau bipolar
y_values = [np.clip(x, -1, 1) for x in x_values]

# Plotar o gráfico
plt.figure(figsize=(10,6))
plt.plot(x_values, y_values, label=title)
plt.title(f'Gráfico da {title}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

# 6 - Função Tangente Hiperbólica

```python
# Função Tangente Hiperbólica
import math
import numpy as np
import matplotlib.pyplot as plt

def hyperbolic_tangent(x, b):
        exponent_term = math.exp(-(b * x))
        return (1 - exponent_term) / (1 + exponent_term)

# Gera valores de x de -5.0 a 5.0 com passo de 0.1
x_values_tanh = np.linspace(-5.0, 5.0, 101)

# Plota o gráfico
plt.figure(figsize=(12, 7))
plt.title('Gráfico da Função Tangente Hiperbólica com Diferentes Valores de b')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, alpha=0.2)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.ylim(-1.1, 1.1)
plt.xticks(np.arange(-5, 6, 1))

# Betas e suas cores correspondentes
b_values_to_plot = [0.5, 1.0, 2.0, 5.0]
colors = ['blue', 'green', 'yellow', 'red']

for beta, color in zip(b_values_to_plot, colors):
    y_values_current_beta = [hyperbolic_tangent(x, beta) for x in x_values_tanh]
    plt.plot(x_values_tanh, y_values_current_beta, color=color, label=f'b = {beta}')

# tanh(x) padrão (tracejado)
plt.plot(x_values_tanh, np.tanh(x_values_tanh), 'k--', label='tanh(x)')

plt.legend()
plt.show()
```

# 7 - Função Gaussiana



```python
# Função Gaussiana
import numpy as np
import matplotlib.pyplot as plt

def gaussian_function(x, sigma):
    return np.exp(-(x**2) / (2 * sigma**2))

x_values_gaussian = np.linspace(-5.0, 5.0, 101)

# Plota o gráfico com múltiplas curvas para diferentes sigmas
plt.figure(figsize=(10, 6))
plt.title('Gráfico da Função Gaussiana com Diferentes Valores de Sigma')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, alpha=0.2)
plt.xticks(np.arange(-5, 5.1, 1))
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.ylim(0, 1.1) # Ajusta o limite Y para melhor visualização

# Sigmas e suas cores
sigma_values_to_plot = [0.5, 1.0, 2.0, 3.0]
colors = ['blue', 'green', 'yellow', 'red']

for sigma, color in zip(sigma_values_to_plot, colors):
    y_values_current_sigma = gaussian_function(x_values_gaussian, sigma)
    plt.plot(x_values_gaussian, y_values_current_sigma, color=color, label=f'σ = {sigma}')

plt.legend()
plt.show()
```

# 8 - Função Ativação Linear


```python
# Função Ativação Linear
import numpy as np
import matplotlib.pyplot as plt

def linear_activation(x):
    return x

# Gera valores de x
x_values_linear = np.linspace(-5.0, 5.0, 101)

# Calcula os valores de y usando a função de ativação linear
y_values_linear = [linear_activation(x) for x in x_values_linear]

# Plota o gráfico
plt.figure(figsize=(10, 6))
plt.plot(x_values_linear, y_values_linear, label='f(x) = x (Função Linear)')
plt.title('Gráfico da Função de Ativação Linear')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, alpha=0.2)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xticks(np.arange(-5, 6, 1)) # Mostra ticks de 1 em 1 no eixo x
plt.yticks(np.arange(-5, 6, 1)) # Mostra ticks de 1 em 1 no eixo y
plt.legend()
plt.show()
```
